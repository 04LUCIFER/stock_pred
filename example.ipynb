```python
from knn_mom import metrics as mc
```

downtrend(df): This function takes in a dataframe df containing stock market data and returns a list of the stock names that are expected to have a downtrend (i.e., the expected percentage change in the stock price is less than -1).

```python
mc.downtrend(ok)
```

uptrend(df): This function takes in a dataframe df containing stock market data and returns a list of the stock names that are expected to have an uptrend (i.e., the expected percentage change in the stock price is greater than 1).

```python
mc.uptrend(ok)
```

up_reversal(df): This function takes in a dataframe df containing stock market data and returns a dataframe containing the stocks that are expected to be in a reversal zone (i.e., expected to have an uptrend). The reversal zone is defined as having a Relative Strength Index (RSI) less than 40 and a Stochastic oscillator (stoch) less than 30.

```python
mc.up_reversal(ok)
```

down_reversal(df): This function takes in a dataframe df containing stock market data and returns a dataframe containing the stocks that are expected to be in a reversal zone (i.e., expected to have a downtrend). The reversal zone is defined as having an RSI greater than 60 and a stoch greater than 80

```python
mc.down_reversal(ok)
```

list_pred(df): This function takes in a dataframe df containing stock market data and returns a dataframe containing the predicted percentage change in the stock price for each stock, along with the stock name. This function can be used to generate a list of predicted closing prices on the current day.

```python
mc.list_pred(ok)
```
